**Question 2) What are the key assumptions of Simple Linear Regression?**

**Answer 2)The key assumptions of Simple Linear Regression are:**

**1) Linearity ‚Äì** The relationship between the independent variable (X) and the dependent variable (Y) is assumed to be linear (a straight-line relationship).

**2) Independence of errors ‚Äì** The residuals (errors) are independent of each other; one observation‚Äôs error does not depend on another‚Äôs.

**3) Homoscedasticity ‚Äì** The variance of the residuals (errors) is constant across all values of the independent variable.

**4) Normality of errors ‚Äì** The residuals are assumed to be normally distributed, especially important for hypothesis testing and confidence intervals.

**5) No or minimal multicollinearity ‚Äì** Since simple linear regression has only one independent variable, this assumption is not usually relevant, but in general regression, independent variables should not be highly correlated.

**Question 3) What is heteroscedasticity, and why is it important to address in regression
models?**

**Answer 3)** Heteroscedasticity refers to a situation in regression analysis where the variance of the residuals (errors) is not constant across all levels of the independent variable. In other words, as the value of the predictor (X) changes, the spread of the errors becomes larger or smaller instead of staying the same. Ideally, regression assumes homoscedasticity, meaning the residuals have equal variance, but when this assumption is violated, it is called heteroscedasticity.

It is important to address heteroscedasticity because it can lead to inefficient estimates of the regression coefficients and make the results of hypothesis tests (like t-tests and F-tests) unreliable. Although heteroscedasticity does not bias the regression coefficients themselves, it affects the accuracy of standard errors, confidence intervals, and significance testing. This may lead to drawing wrong conclusions about the relationship between variables.

```python
#Question 6)  Implement a Python program to fit a Simple Linear Regression model to
#the following sample data:
#‚óè X = [1, 2, 3, 4, 5]
#‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2]
#Plot the regression line over the data points.


import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

print("Intercept (Œ≤0):", model.intercept_)
print("Slope (Œ≤1):", model.coef_[0])
print("Regression Equation: Y =", model.intercept_, "+", model.coef_[0], "* X")

# Plot data points
plt.scatter(X, Y, color='blue', label="Actual Data")

# Plot regression line
plt.plot(X, Y_pred, color='red', linewidth=2, label="Regression Line")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression Example")
plt.legend()
plt.show()
```

```python
# Question 7) Fit a Multiple Linear Regression model on this sample data:
#‚óè Area = [1200, 1500, 1800, 2000]
#‚óè Rooms = [2, 3, 3, 4]
#‚óè Price = [250000, 300000, 320000, 370000]
#Check for multicollinearity using VIF and report the results.


import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample data
data = {
    "Area": [1200, 1500, 1800, 2000],
    "Rooms": [2, 3, 3, 4],
    "Price": [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Independent variables (X) and dependent variable (Y)
X = df[["Area", "Rooms"]]
Y = df["Price"]

# Fit Multiple Linear Regression
model = LinearRegression()
model.fit(X, Y)

X_with_const = sm.add_constant(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i)
                   for i in range(X_with_const.shape[1])]


print("Intercept (Œ≤0):", model.intercept_)
print("Coefficients (Œ≤):", dict(zip(X.columns, model.coef_)))
print("\nVariance Inflation Factor (VIF):")
print(vif_data)
```

```python
# Question 8) Implement polynomial regression on the following data:
#‚óè X = [1, 2, 3, 4, 5]
#‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]
#Fit a 2nd-degree polynomial and plot the resulting curve.


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])


poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values for smooth curve
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
Y_pred = model.predict(poly.transform(X_range))

# Print regression equation
print("Intercept (Œ≤0):", model.intercept_)
print("Coefficients (Œ≤):", model.coef_)
print(f"Equation: Y = {model.intercept_:.2f} + {model.coef_[1]:.2f}*X + {model.coef_[2]:.2f}*X^2")


plt.scatter(X, Y, color='blue', label="Actual Data")
plt.plot(X_range, Y_pred, color='red', linewidth=2, label="Polynomial Fit (Degree 2)")

# Labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")
plt.legend()
plt.show()

```

```python
#Question 9) Create a residuals plot for a regression model trained on this data:
#‚óè X = [10, 20, 30, 40, 50]
#‚óè Y = [15, 35, 40, 50, 65]
#Assess heteroscedasticity by examining the spread of residuals.


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Print regression equation and residuals
print("Intercept (Œ≤0):", model.intercept_)
print("Slope (Œ≤1):", model.coef_[0])
print("Residuals:", residuals)

# Residuals plot
plt.scatter(X, residuals, color='blue', label="Residuals")
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.legend()
plt.show()
```

**Question 10) Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.**


**Answer 10) Addressing Heteroscedasticity**

Heteroscedasticity means the variance of residuals is not constant. To fix it:

‚úÖ Transform the dependent variable ‚Äì Apply log, square root, or Box-Cox transformation on house prices to stabilize variance.

‚úÖ Use Weighted Least Squares (WLS) ‚Äì Give less weight to data points with higher variance, improving model efficiency.

‚úÖ Use robust standard errors ‚Äì This ensures valid hypothesis testing even when heteroscedasticity is present.

‚úÖ Feature engineering ‚Äì Sometimes adding interaction terms or polynomial features can capture the variance better.

**Addressing Multicollinearity**

Multicollinearity means predictors (e.g., area and number of rooms) are highly correlated. To fix it:

‚úÖ Check VIF (Variance Inflation Factor) ‚Äì Identify variables causing multicollinearity.

‚úÖ Remove or combine correlated variables ‚Äì For example, instead of using both area and number of rooms, create a new feature like area per room.

‚úÖ Apply dimensionality reduction (PCA) ‚Äì Reduce correlated predictors into uncorrelated components.

‚úÖ Use Regularization (Ridge/Lasso Regression) ‚Äì These methods penalize large coefficients, making the model more stable under multicollinearity.

**Building a Robust Model**

* After addressing both issues, I would retrain the model and check residual plots and VIF values again.

* I would also validate the model using cross-validation to ensure it generalizes well.

* If non-linear relationships are strong, I might even switch to tree-based models (like Random Forests or Gradient Boosting), which handle heteroscedasticity and multicollinearity better than linear regression.
